```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv')
df.rename(columns={col: col.strip() for col in df.columns}, inplace=True)
```

```python
df
```

```python
plt.figure(figsize=(12,8))
plt.plot(df['epoch'], df['metrics/precision(M)'])
plt.plot(df['epoch'], df['metrics/recall(M)'])
plt.plot(df['epoch'], df['metrics/mAP50(M)'])
plt.plot(df['epoch'], df['metrics/mAP50-95(M)'])

plt.grid()
plt.legend(['Precision', 'Recall', 'mAP50', 'mAP50-95'])
plt.title("Performance")
plt.show()
```

```python
def plot_train_val_loss(x, train, val, title):
    plt.figure(figsize=(12,8))
    plt.plot(x, train)
    plt.plot(x, val)
    plt.grid()
    plt.legend(['train', 'val'], loc='best')
    plt.title(title)
    plt.show()

plot_train_val_loss(df['epoch'], df['train/box_loss'], df['val/box_loss'], "Box Loss")
```

```python
plot_train_val_loss(df['epoch'], df['train/seg_loss'], df['val/seg_loss'], "Seg Loss")
```

```python
plot_train_val_loss(df['epoch'], df['train/cls_loss'], df['val/cls_loss'], "Cls Loss")
```

```python
plot_train_val_loss(df['epoch'], df['train/dfl_loss'], df['val/dfl_loss'], "DFL Loss")
```
